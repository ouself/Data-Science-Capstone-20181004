Use BeautifulSoup to open the link and scrape down the table. And build dataframe from the scraped table while giving new column names. Also, delet data that has'Borough' as 'Not assigned', and copy 'Neighborhood' to the same value of 'Bourough' if 'Neighborhood' 's value is 'Not assigned'. 
Finally, group the value belonging to the same postal code of neighbohood in one column and remove /n of neighbohood column.

```python
sauce=urllib.request.urlopen('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').read()
soup = bs.BeautifulSoup(sauce, 'lxml')
table = soup.table
table = soup.find('table')
table_rows = table.find_all('tr')
l = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text for tr in td]
    l.append(row)
a = pd.DataFrame(l, columns=["Postal Code", "Borough", 'Neighborhood'])
a = a.drop(a[a.Borough == 'Not assigned'].index)
b = a.replace('\n','', regex=True)
c = b[1:]
c.loc[c['Neighborhood'] == 'Not assigned', 'Neighborhood'] = c['Borough']
d = c.groupby('Postal Code', as_index=False).agg(lambda x: ', '.join(set(x.astype(str))))
d
```
